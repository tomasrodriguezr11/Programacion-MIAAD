# Objetivo:
Utilizar GridSpec para organizar subplots de diferentes tamaños en un análisis de ventas.

# Instrucciones:
1. Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros. Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.
2. Usa GridSpec para organizar los subplots de la siguiente manera:
- Primer gráfico (barra, ocupando dos columnas)
- Segundo gráfico (línea, una columna)
- Tercer gráfico (dispersión, dos filas).
3. Agrega títulos y ajusta espacios.

Sube un archivo txt que contenga el enlace a tu notebook.

```python
#Importación de librerías que se usarán en esta práctica
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
# 1. Creación de DataFrame de ventas con Mes, Ventas y Gastos
np.random.seed(42)

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun",
         "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

df = pd.DataFrame({
    "Mes": meses,
    "Ventas": np.random.randint(2000, 5001, 12),
    "Gastos": np.random.randint(1000, 4001, 12)
})

df
```

```python
# 2. Se usa GridSpec para organizar los subplots
fig = plt.figure(figsize=(14, 8)) #Crear figura
gs = GridSpec(2, 3, figure=fig) #Definir una cuadrícula de 2x3

# Primer gráfico (barra, ocupando dos columnas)
ax1 = fig.add_subplot(gs[0, 0:2]) #Fila 0, columnas 0 a 1
ax1.bar(df["Mes"], df["Ventas"], edgecolor='dimgray')
ax1.set_title("Gráfico de barras de Ventas Mensuales", fontsize=14)
ax1.set_ylabel("Ventas")
ax1.tick_params(axis='x', rotation=45)

# Segundo gráfico (línea, una columna)
ax2 = fig.add_subplot(gs[0, 2]) #Fila 0, columna 2
ax2.plot(df["Mes"], df["Gastos"], marker='o')
ax2.set_title("Gráfico de líneas de Gastos Mensuales", fontsize=12)
ax2.set_ylabel("Gastos")
ax2.tick_params(axis='x', rotation=45)

# Tercer gráfico (dispersión, dos filas).
ax3 = fig.add_subplot(gs[1, :]) #Ocupa toda la segunda y tercera fila
ax3.scatter(df["Ventas"], df["Gastos"])
ax3.set_title("Gráfico de dispersión - Relación Ventas vs Gastos", fontsize=14)
ax3.set_xlabel("Ventas")
ax3.set_ylabel("Gastos")


# 3. Ajuste de espacios
plt.tight_layout()
plt.show()
```
