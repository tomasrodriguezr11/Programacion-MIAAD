- (0,0): sns.histplot(): Histograma de las longitudes de aletas.

```python
#Importación de librerías que se usarán en esta práctica
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Carga del dataset "penguins" de seaborn
df = sns.load_dataset("penguins")
df = df.dropna() #Se utiliza para evitar valores nulos
df
```

```python
# 2. Usa plt.subplots() para crear una figura con 2 filas y 2 columnas:

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# 3. Genera los siguientes gráficos y personalízalos:

# (0,0): sns.histplot(): Histograma de las longitudes de aletas.
sns.histplot(
    data=df,
    x="flipper_length_mm",
    bins=20,
    kde=True,
    color="skyblue",
    ax=ax[0, 0]
)

ax[0, 0].set_title("Distribución de Longitud de Aletas")
ax[0, 0].set_xlabel("Longitud de Aleta (mm)")

# (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
sns.boxplot(
    data=df,
    x="species",
    y="body_mass_g",
    palette="pastel",
    ax=ax[0, 1]
)

ax[0, 1].set_title("Masa Corporal por Especie")
ax[0, 1].set_xlabel("Especie")
ax[0, 1].set_ylabel("Masa Corporal (g)")

# (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
sns.scatterplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    palette="deep",
    ax=ax[1, 0]
)

ax[1, 0].set_title("Masa Corporal vs Longitud de Aleta")
ax[1, 0].set_xlabel("Longitud de Aleta (mm)")
ax[1, 0].set_ylabel("Masa Corporal (g)")

# (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
sns.violinplot(
    data=df,
    x="species",
    y="body_mass_g",
    palette="muted",
    ax=ax[1, 1]
)

ax[1, 1].set_title("Distribución de Masa Corporal por Especie")
ax[1, 1].set_xlabel("Especie")
ax[1, 1].set_ylabel("Masa Corporal (g)")

# 5. Ajuste del diseño
plt.tight_layout()
plt.show()
```
