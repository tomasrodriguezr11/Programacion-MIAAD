# **Práctica 12: Carga de Datos**

```python
import pandas as pd

# Carga del archivo miaad-nyc-r-s a un dataframe omitiendo la última fila
dataset="drive/MyDrive/ProgramacionMIAAD/miaad-nyc-r-s.csv"
df = pd.read_csv(dataset, skipfooter=1, engine='python')
df.tail(3)
```

```python
# Cantidad de filas y columnas del dataframe
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])
```

# **Práctica 13: Cambiar nombre de columna**

```python
# Nombres de la columnas del archivo
print("Nombre de columnas:")
print(df.columns)
```

```python
# La primera columna no se llama INDEX MIAAD

# Cambio de nombre de la primera columna a INDEX MIAAD
df.rename(columns={df.columns[0]: "INDEX MIAAD"}, inplace=True)

print("Nombre de columnas después del cambio:")
print(df.columns)
```

#**Práctica 14: Normalización de los nombres de la columna.**

```python
# Normalización de los nombres de las columnas en minúsculas y cambiando los espacios por guion bajo
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Nombres de las columnas normalizadas
print("Columnas normalizadas:")
print(df.columns)
```

#**Práctica 15: Errores en los Tipos de Datos**

```python
# Tipos de datos de cada columna.
print("Tipos de datos:")
print(df.dtypes)
```

```python
# Tipos de datos de las columnas SALE PRICE(sale_price), LAND SQUARE FEET (land_square_feet), GROSS SQUARE FEET (gross_square_feet) - Deben ser tipo flotante.
print("\nTipos de datos de las siguientes columnas:")
print("sale_price:", df["sale_price"].dtype)
print("land_square_feet:", df["land_square_feet"].dtype)
print("gross_square_feet:", df["gross_square_feet"].dtype)
```

```python
# Debido a que las columnas tienen tipo de dato "object" y no flotante, se realiza el cambio
df["sale_price"] = pd.to_numeric(df["sale_price"], errors="coerce")
df["land_square_feet"] = pd.to_numeric(df["land_square_feet"], errors="coerce")
df["gross_square_feet"] = pd.to_numeric(df["gross_square_feet"], errors="coerce")

# Tipos de datos de las columnas con el ajuste.
print("\nTipos de datos después de la corrección:")
print("sale_price:", df["sale_price"].dtype)
print("land_square_feet:", df["land_square_feet"].dtype)
print("gross_square_feet:", df["gross_square_feet"].dtype)
```

#**Práctica 16: Diccionario de datos**

```python
# Creación del diccionario de datos del dataframe preprocesado. Se considera tipo de dato y cantidad de valores
diccionario_datos = {}

for columna in df.columns:
    diccionario_datos[columna] = {
        "tipo_de_dato": str(df[columna].dtype),
        "cantidad_de_valores": df[columna].count(),
    }

# Contenido del diccionario
diccionario_datos
```
