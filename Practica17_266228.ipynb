```python
# Carga de liberías y dataset
import pandas as pd

dataset="drive/MyDrive/ProgramacionMIAAD/nls97.csv"
df = pd.read_csv(dataset)
df
```

# **Ejercicio 01: Errores Tipográficos y Valores Inconsistentes: maritalstatus**

```python
# Valores únicos de la columna maritalstatus
display(df['maritalstatus'].unique())
```

```python
# Frecuencia de cada valor único de la columna maritalstatus
display(df['maritalstatus'].value_counts())
```

```python
# Cambio del valor 'Never-Married' a 'Single'
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')

# Frecuencia de cada valor úncio después del cambio
display(df['maritalstatus'].value_counts())
```

# **Ejercicio 02: Errores Tipográficos y Valores Inconsistentes: highestdegree**

```python
# Valores únicos de la columna highestdegree
display(df['highestdegree'].unique())
```

```python
# Frecuencia de cada valor único de la columna highestdegree
display(df['highestdegree'].value_counts())
```

```python
# Modificación de categorías: Convertir a minúsculas y quitar número
df['highestdegree'] = (
    df['highestdegree']
    .str.replace(r'^\d+\.\s*', '', regex=True)
    .str.lower()
)

# Frecuencia de cada valor úncio después del cambio
display(df['highestdegree'].unique())
```

#**Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas**

```python
# Unión de las categorías Widowed y Single de la columna maritalstatus en una categoría más general llamada Single/No Partner
# Creación de nueva columna marital_grouped para las nuevas categorías de maritalstatu

df['marital_grouped'] = df['maritalstatus'].replace({
    'Widowed': 'Single/No Partner',
    'Single': 'Single/No Partner'
})

# Valores únicos de la nueva columna marital_gruped
display(df['marital_grouped'].unique())
```

```python
# Frecuencia de cada valor único de la nueva columna marital_gruped
display(df['marital_grouped'].value_counts())
```

#**Ejercicio 04: Tipos de Datos**

```python
# Tipo de datos de la columna gender
print(df['gender'].dtype)
```

```python
# Cantidad de bytes de columna gender
print(df['gender'].memory_usage(deep=True))
```

```python
# Columna tipo object se modifica a tipo category
if df['gender'].dtype == 'object':
    df['gender'] = df['gender'].astype('category')

# Cantidad de bytes que ocupa la columna gender con el nuevo tipo de datos
print(df['gender'].memory_usage(deep=True))
```
