# **Ejercicio 01: Tipo de Datos Correcto**

```python
# Carga de liberías y conjunto de datos
import pandas as pd

dataset="drive/MyDrive/ProgramacionMIAAD/Airdata.csv"
df = pd.read_csv(dataset)
df
```

```python
# Revisión de tipo de datos por columna
print(df.dtypes)
```

```python
# Conversión de columna DateTime a tipo temporal si es necesario
df['DateTime'] = pd.to_datetime(df['DateTime'], errors='coerce')

print("\nTipos de datos después de conversión:")
print(df.dtypes)
```

```python
#Impresión de las primeras tres líneas
print("\nPrimeras 3 filas:")
print(df.head(3))
```

# **Ejercicio 02: Extraer componentes de fecha y tiempo**

```python
# Extracción de año, mes, día y hora de la columna 'DateTime' en nueva columnas para cada componente
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Hour'] = df['DateTime'].dt.hour

# Impresión de las primeras tres líneas del dataframe
print("\nPrimeras 3 filas con componentes extraídos:")
print(df.head(3))
```

# **Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)**

```python
# Se suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime' y se guarda el resultado en una nueva columna para el dataframe.
df['DateTime_suma'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)

# Impresión de la columna 'DateTime' y de la columna que resulta de sumar 10 días, 7 horas y 15 minutos.
print("\nColumnas DateTime y DateTime_suma:")
print(df[['DateTime', 'DateTime_suma']].head())
```

# **Ejercicio 04: Timestamp**

```python
# Conversión de la columna 'DateTime' a timestamp Unix
df['Unix_Timestamp'] = df['DateTime'].astype('int64') // 10**9

# Impresión de las primeras tres filas de  la columna 'DateTime' y la columna que representa dicha fecha en timestamp Unix
print("\nPrimeras 3 filas con Timestamp Unix:")
print(df[['DateTime', 'Unix_Timestamp']].head(3))
```

# **Ejercicio 05: Filtrado de Fechas**

```python
# Se filtran las fechas que son mayores a 2020-11-01 en la columna 'DateTime'
filtro_df = df[df['DateTime'] > pd.to_datetime('2020-11-01')]

print("\nFechas mayores a 2020-11-01:")
print(filtro_df)
```
