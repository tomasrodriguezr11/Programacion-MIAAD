# **Ejercicio 01: Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration**

```python
# Carga de liberías y conjunto de datos
import pandas as pd

dataset="drive/MyDrive/ProgramacionMIAAD/dirtydata.csv"
df = pd.read_csv(dataset)
df
```

```python
# Conteo de registros duplicados usando solo la columna Duration
duplicados_duration = df.duplicated(subset=['Duration']).sum()
print("Cantidad de registros duplicados en Duration:", duplicados_duration)
```

```python
# Cantidad de registros que hay por cada valor único con value_counts
print("\nConteo de valores únicos en Duration:")
print(df['Duration'].value_counts())
```

```python
# Se filtran los registros sin duplicados
df_sin_dup_duration = df.drop_duplicates(subset=['Duration'])

print("\nDatos sin duplicados en Duration:")
df_sin_dup_duration
```

# **Ejercicio 02: Duplicidad Parcial Dos columnaa: Contar y Filtrar los registros duplicados en Pulse y MaxPulse**

```python
# Conteo de registros duplicados de las columnas Pulse y Maxpulse
duplicados_pulse = df.duplicated(subset=['Pulse', 'Maxpulse']).sum()
print("Cantidad de registros duplicados en Pulse y Maxpulse:", duplicados_pulse)
```

```python
# Cantidad de registros que hay por cada valor único usando value_counts()
print("\nConteo de combinaciones Pulse - Maxpulse:")
print(df.value_counts(subset=['Pulse', 'Maxpulse']))
```

```python
# Se filtran los datos y se obtiene los registros cuando keep = 'last'
df_sin_dup_pulse = df.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')

print("\nDatos sin duplicados (conservando última ocurrencia):")
print(df_sin_dup_pulse)
```

# **Ejercicio 03: Duplicidad Total o Exacta: Contar y Filtrar los registros duplicados**

```python
# Cantidad de registros duplicados exactos (todas las columnas) utilizando sum()
duplicados_exactos = df.duplicated().sum()

print("Cantidad de registros con duplicidad exacta:", duplicados_exactos)
```

```python
# Filtra los datos y obten los registros cuando keep = 'last'
df_sin_dup_exactos = df.drop_duplicates(keep='last')

print("\nDatos sin duplicados exactos:")
print(df_sin_dup_exactos)
```

# **Ejercicio 04: Cálculo de la media con duplicados y sin duplicados**

```python
# Cálculo de la media de las columnas Pulse y Maxpulse
media_con_dup = df[['Pulse', 'Maxpulse']].mean()
print("\nMedia con duplicados:")
print(media_con_dup)
```

```python
# Identificación de valores duplicados en las columnas Pulse y Maxpulse (Duplicidad Parcial) y eliminación de registros repetidos (quedándose con las últimas ocurrencias)
df_sin_dup = df.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')
df_sin_dup
```

```python
# Cálculo de la media de las columnas Pulse y Maxpulse utilizando el conjunto anterior sin datos duplicados (donde solo quedaron las últimas ocurrencias de los datos)
media_sin_dup = df_sin_dup[['Pulse', 'Maxpulse']].mean()

print("\nMedia sin duplicados:")
print(media_sin_dup)
```
