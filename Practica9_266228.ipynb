# Objetivo:
Utilizar plt.subplots() para visualizar diferentes distribuciones generadas con NumPy y almacenadas en un DataFrame.

# Instrucciones:

- Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).
- Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.
- Grafica cada distribución en un subplot usando ax[i, j].hist().
- Agrega títulos a cada subplot y ajusta la disposición con tight_layout().

Sube un archivo txt que contenga el enlace a tu notebook.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson)
np.random.seed(42)  # para que los resultados sean reproducibles

df = pd.DataFrame({
    "Normal": np.random.normal(loc=0, scale=1, size=1000),
    "Uniforme": np.random.uniform(low=0, high=1, size=1000),
    "Exponencial": np.random.exponential(scale=1, size=1000),
    "Poisson": np.random.poisson(lam=3, size=1000)
})

df
```

```python
# 2. Figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# 3. Gráfica de cada distribución en un subplot usando ax[i, j].hist() con títulos a cada subplot
ax[0, 0].hist(df["Normal"], bins=30, color="b", edgecolor='black', linewidth=1)
ax[0, 0].set_title("Distribución Normal")

ax[0, 1].hist(df["Uniforme"], bins=30, color="r", edgecolor='black', linewidth=1)
ax[0, 1].set_title("Distribución Uniforme")

ax[1, 0].hist(df["Exponencial"], bins=30, color="g", edgecolor='black', linewidth=1)
ax[1, 0].set_title("Distribución Exponencial")

ax[1, 1].hist(df["Poisson"], bins=30, color="purple", edgecolor='black', linewidth=1)
ax[1, 1].set_title("Distribución Poisson")


# 4. Ajuste de la disposición con tight_layout().
plt.tight_layout()
plt.show()
```
