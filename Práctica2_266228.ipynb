# **Explicación breve del código:**

Este programa abre el archivo informe_acciones.txt y analiza cada una de sus líneas.
Haciendo uso de una expresión regular, se identifican las líneas que contienen el texto "Valor actual:" seguido de un símbolo de dólar y un número.
Se extrae el valor numérico de cada acción, se convierte a tipo float y después se almacena en una lista.
Finalmente, el programa imprime en pantalla todos los valores de las acciones encontrados en los correos electrónicos.

```python
#Importamos la biblioteca re para trabajar con expresiones regulars
#en Python
import re

# Abrimos el archivo
#Leemos el archivo ubicado en Google Drive
#El archivo se abre en modo lectura por defecto
archivo = open('/content/drive/MyDrive/Programación MIAAD/informe_acciones.txt', encoding='latin-1')

# Lista para guardar los valores de las acciones
valores_acciones = []

# Expresión regular para extraer el valor numérico después del símbolo $
patron = r'Valor actual:\s*\$([0-9]+(?:\.[0-9]+)?)'
#\s*	Permite espacios
#\$	Detecta el símbolo $
#[0-9]+	Parte entera del número
#(?:\.[0-9]+)?	Decimales (opcional)

# Recorremos cada línea del archivo
for linea in archivo:
    resultado = re.findall(patron, linea)
    if resultado:
        valores_acciones.append(float(resultado[0]))

# Cerramos el archivo
archivo.close()

# Mostramos los valores extraídos
for valor in valores_acciones:
    print(valor)
```
