# Instrucciones

Crea un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100. Luego, realiza las siguientes operaciones:

1. Encuentra el valor máximo del arreglo.

2. Encuentra el valor mínimo del arreglo.

3. Calcula la media del arreglo.

4. Suma 10 a cada elemento del arreglo.

No olvides enviar el archivo TXT con las ligas a tu Notebook

```python
import numpy as np
```

```python
# Se crea un arreglo unidimensional de 20 elementos aleatorios entre 1 y 100
arreglo = np.random.randint(1, 101, size=20)
print(arreglo)
```

```python
# 1. Encontrar el valor máximo del arreglo
valor_max = np.max(arreglo)
print(valor_max)
```

```python
# 2. Encontrar el valor mínimo del arreglo
valor_min = np.min(arreglo)
print(valor_min)
```

```python
# 3. Calcular la media del arreglo
media = np.mean(arreglo)
print(media)
```

```python
# 4. Suma 10 a cada elemento del arreglo
arreglo_con_10 = arreglo + 10
print(arreglo_con_10)
```
