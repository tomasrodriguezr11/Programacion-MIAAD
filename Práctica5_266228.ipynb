# Instrucciones

Uso de máscaras booleanas.

Crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50. Luego:

1. Selecciona los elementos mayores que 25 utilizando una máscara booleana.
2. Cuenta cuántos elementos son mayores que 25.


No olvides enviar el archivo TXT con la liga a tu notebook en Google Colab y Github

```python
import numpy as np
```

```python
# Se crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50
arreglo = np.random.randint(1, 51, size=15)
print(arreglo)
```

```python
# 0. Creación de la máscara booleana (elementos mayores que 25)
mascara = arreglo > 25
print(mascara)
```

```python
# 1. Selección de los elementos mayores que 25 usando la máscara
mayores_25 = arreglo[mascara]
print(mayores_25)
```

```python
# 2. Conteo de cuántos elementos son mayores que 25
cantidad = np.sum(mascara)
print(cantidad)
```
