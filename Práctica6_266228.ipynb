```python
import pandas as pd
import numpy as np

# Carga el archivo CSV en un DataFrame de pandas
df_titanic = pd.read_csv('/content/drive/MyDrive/Programación MIAAD/Titanic-Dataset.csv')

# Imprime las primeras filas del DataFrame para visualizarlo como tabla
display(df_titanic.head())
```

```python
# Ejercicio 1: Análisis de la distribución de supervivencia por combinación de sexo y clase del pasajero.

# Calcula la proporción de supervivencia para cada combinación de 'Sex' y 'Pclass'.
supervivencia_por_grupo = (
    df_titanic
    .groupby(['Sex', 'Pclass'])['Survived']
    .mean()
)

supervivencia_por_grupo

# Identifica qué combinación tuvo la tasa de supervivencia más alta.
supervivencia_alta = supervivencia_por_grupo.idxmax(), supervivencia_por_grupo.max()
print("Mayor tasa de supervivencia:", supervivencia_alta)

# Identifica qué combinación tuvo la tasa de supervivencia más baja.
supervivencia_baja = supervivencia_por_grupo.idxmin(), supervivencia_por_grupo.min()
print("Menor tasa de supervivencia:", supervivencia_baja)
```

```python
# Ejercicio 2: Identificación de familias grandes a bordo.

# Crea una nueva columna 'FamilySize' sumando las columnas 'SibSp' y 'Parch'.
df_titanic['FamilySize'] = df_titanic['SibSp'] + df_titanic['Parch']

# Considera como "familia grande" a aquellas donde 'FamilySize' es mayor a 3.
familias_grandes = df_titanic[df_titanic['FamilySize'] > 3]

# Calcula el número de pasajeros en familias grandes.
num_familias_grandes = len(familias_grandes)

# Calcula la proporción de supervivencia entre los pasajeros que pertenecen a familias grandes.
survival_familias_grandes = familias_grandes['Survived'].mean()

print("Número de pasajeros en familias grandes:", num_familias_grandes)
print("Proporción de supervivencia entre los pasajeros que pertenecen a familias grandes:", survival_familias_grandes)
```

```python
# Ejercicio 3: Segmentación por grupos de edad.

# Clasifica a los pasajeros en las siguientes categorías de edad:  menor de edad (< 18) y mayor de edad (>=18)
def clasificar_edad(edad):
    if pd.isna(edad):
        return np.nan
    elif edad < 18:
        return "Menor de edad"
    else:
        return "Mayor de edad"
df_titanic['GrupoEdad'] = df_titanic['Age'].apply(clasificar_edad)

df_titanic[['Age', 'GrupoEdad']].head()
```

```python
# Ejercicio 4: Comparación entre promedios calculados manualmente y con Pandas

# Utiliza NumPy para calcular el promedio de las columnas 'Age' y 'Fare', ignorando valores nulos.
edad_np = np.nanmean(df_titanic['Age'].values)
fare_np = np.nanmean(df_titanic['Fare'].values)

print(f"Edad con NumPy: {edad_np:.2f}")
print(f"Fare con NumPy: {fare_np:.2f}")
print()

# Compara estos valores con los promedios obtenidos utilizando los métodos nativos de Pandas.
edad_pd = df_titanic['Age'].mean()
fare_pd = df_titanic['Fare'].mean()

print(f"Edad con Pandas: {edad_pd:.2f}")
print(f"Fare con Pandas: {fare_pd:.2f}")

# Verifica que los resultados sean consistentes.
```

```python
# Ejercicio 5. Creación de intervalos de clase usando NumPy y análisis con Pandas

# Divide la columna 'Fare' en 5 intervalos equidistantes utilizando la función numpy.linspace.
intervalos = np.linspace(df_titanic['Fare'].min(), df_titanic['Fare'].max(), 6)
intervalos

# Crea una nueva columna en el DataFrame que asigne a cada pasajero el intervalo correspondiente de su tarifa.
df_titanic['FareInterval'] = pd.cut(df_titanic['Fare'], bins=intervalos, include_lowest=True)
df_titanic[['Fare', 'FareInterval']].head()

# Calcula el número de pasajeros en cada intervalo utilizando Pandas y la proporción de supervivientes por intervalo.
analisis_fare = (
    df_titanic
    .groupby('FareInterval')['Survived']
    .agg(
        pasajeros='count',
        proporcion_supervivencia='mean'
    )
)

analisis_fare
```
