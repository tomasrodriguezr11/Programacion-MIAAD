```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carga y exploración del dataset:**

# Cargar el dataset Iris utilizando seaborn o cualquier otro dataset de su elección.
df = sns.load_dataset('iris')

# Verificar las primeras filas del dataset con el método .head().
df.head()
```

```python
# Obtener información general sobre los datos (nombres de columnas, tipos de datos, valores nulos, etc.).
df.info()
```

```python
# a) Gráfico de barras:

# Cálculo de promedios
promedios = df.groupby('species')[['petal_length', 'petal_width']].mean()

# Creación de gráfico de barras
promedios.plot(kind='bar', figsize=(8, 5))

plt.title('Promedio del largo y ancho del pétalo por especies de Iris')
plt.xlabel('Especie')
plt.ylabel('Medida o promedio (cm)')
plt.legend(title='Variable')
plt.tight_layout()
plt.show()
```

```python
# b) Histograma:

# Creación de histograma
plt.figure(figsize=(8, 5))
plt.hist(df['sepal_length'], bins=20, color='skyblue', edgecolor='gray')

plt.title('Distribución del largo del sépalo')
plt.xlabel('Largo del sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

```python

# c) Gráfico de dispersión (scatter plot):

# Creación de gráfico de dispersión
plt.figure(figsize=(8, 5))

for especie in df['species'].unique():
    subset = df[df['species'] == especie]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=especie)

plt.title('Relación entre largo y ancho del pétalo')
plt.xlabel('Largo del pétalo (cm)')
plt.ylabel('Ancho del pétalo (cm)')
plt.legend(title='Especie')
plt.show()
```

```python
# d) Gráfico de cajas (box plot):

# Creación de gráfico de cajas - se usan subplots para mejorar la visualización
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=False)

# Gráfico de cajas para sepal_length
sns.boxplot(x='species', y='sepal_length', data=df, ax=axes[0])
axes[0].set_title('Distribución del largo del sépalo por especie')
axes[0].set_xlabel('Especie')
axes[0].set_ylabel('Largo del sépalo (cm)')

# Gráfico de cajas para sepal_width
sns.boxplot(x='species', y='sepal_width', data=df, ax=axes[1])
axes[1].set_title('Distribución del ancho del sépalo por especie')
axes[1].set_xlabel('Especie')
axes[1].set_ylabel('Ancho del sépalo (cm)')

plt.suptitle('Distribución de las medidas del sépalo por especie', y=1.02)
plt.tight_layout() # Ajustar el diseño para evitar superposiciones
plt.show()
```

```python
# e) Gráfico de líneas:

# Cración del gráfico de líneas
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['petal_length'])

plt.title('Tendencia de la longitud del pétalo')
plt.xlabel('Índice de observación')
plt.ylabel('Largo del pétalo (cm)')
plt.show()
```
