```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carga y exploración del dataset:

# Cargar el dataset tips usando seaborn.
df = sns.load_dataset('tips')

# Realizar un análisis preliminar: utiliza .head(), .info(), .describe() para obtener una visión general de los datos.
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Verificar la existencia de valores nulos o datos faltantes.
df.isnull().sum()
```

```python
# a) Mapa de Calor (Heatmap):

# Calcular matriz de correlación
correlacion = df[['total_bill', 'tip', 'size']].corr()

#Creación del mapa de calor
plt.figure(figsize=(6, 5))
sns.heatmap(
    correlacion,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Mapa de calor de correlación entre variables')
plt.show()
```

```python
# b) Diagrama de Violín:

# Creación del diagrama de Violín
plt.figure(figsize=(8, 5))
sns.violinplot(
    x='day',
    y='tip',
    data=df,
    palette='Set2'
)

plt.title('Distribución de las propinas por día')
plt.xlabel('Día de la semana')
plt.ylabel('Propina')
plt.show()
```

```python
# c) Gráfico de dispersión (Scatter Plot):

# Creación del gráfico de dispersión
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='day',
    data=df
)

plt.title('Relación entre total de la cuenta (total_bill) y propina (tip)')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.legend(title='Día')
plt.show()
```

```python
# d) Gráfico de barras (Bar Plot):

#Creación del gráfico de barras
plt.figure(figsize=(6, 5))
sns.barplot(
    x='sex',
    y='tip',
    data=df,
    palette='pastel'
)

plt.title('Promedio de propina por género')
plt.xlabel('Género')
plt.ylabel('Propina promedio')
plt.show()
```
